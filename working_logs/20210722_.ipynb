```python
def convert_pdf_to_txt(pdf_file):
    
    # Create object for recousre manager of pdf
    rsrcmgr = PDFResourceManager()
    
    # Process string data as a file -> Contain PDF contents 
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)
    fp = open(pdf_file, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    # maxpages
    caching = True
    pagenos = set()
    
    for page in PDFPage.get_pages(fp, pagenos, password=password, caching=caching, check_extractable=True):
        interpreter.process_page(page)
    text = retstr.getvalue()
    
    # save text as txt file 
    tmp = text.strip('\n')
    with open (r'C:\Users\MI2RL-KHJ\mfds\product_grade_1.txt','w', encoding='UTF-8') as f: 
        f.write(tmp)
        
    # Processes re1sult in text
    fp.close()
    device.close()
    retstr.close()
    return text
```
